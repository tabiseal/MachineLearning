```python
import numpy as np
import matplotlib.pyplot as plt
import dataset
```

```python
xs,ys=dataset.get_beans(100)

#配置图像
plt.title("Size-Toxicity Function",fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")

plt.scatter(xs,ys)
w=0.1
y_pre=w*xs

plt.plot(xs,y_pre)

plt.show()
```

データ100個場合の合計の誤差：

```python
m=100
es=(ys-y_pre)**2
sum_e=np.sum(es)
sum_e=(1/m)*sum_e
```

wが1から3まで違う値から得られる誤差：

```python
ws=np.arange(0,3,0.1)
es=[]
for w in ws:
    y_pre=w*xs
    e=(1/m)*np.sum((ys-y_pre)**2)
    es.append(e)
    
    
#配置图像
plt.title("Cost Function",fontsize=12)
plt.xlabel("w")
plt.ylabel("e")
plt.plot(ws,es)
plt.show()
```

```python
w_min=np.sum(xs*ys)/np.sum(xs*xs)
print("e最小点w:"+str(w_min))
```

eが最小値の時の予測関数

```python
y_pre=w_min*xs

#配置图像
plt.title("Cost Function",fontsize=12)
plt.xlabel("w")
plt.ylabel("e")

plt.scatter(xs,ys)
plt.plot(xs,y_pre)
plt.show()
```
