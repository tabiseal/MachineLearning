```python
import numpy as np
import matplotlib.pyplot as plt
import dataset
```

```python
xs,ys=dataset.get_beans(100)

#配置图像
plt.title("Size-Toxicity Function",fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")


plt.scatter(xs,ys)

b=0.1
w=0.1

z=w*xs+b
a=1/(1+np.exp(-z))


plt.plot(xs,a)

plt.show()
```

```python
for m in range(500):
    for i in range(100):
        x=xs[i]
        y=ys[i]
        
        #对w和b求偏导
        z=w*x+b
        a=1/(1+np.exp(-z))
        e=(y-a)**2
        
        deda=-2*(y-a)
        dadz=a*(1-a)
        dzdw=x
        
        dedw=deda*dadz*dzdw
        
        dzdb=1
        
        dedb=deda*dadz*dzdb
        
        alpha=0.05
        
        w=w-alpha*dedw
        b=b-alpha*dedb
        
plt.scatter(xs,ys)

z=w*xs+b
a=1/(1+np.exp(-z))

plt.xlim(0,1)
plt.ylim(0,1.2)

plt.plot(xs,a)
plt.show()
```
