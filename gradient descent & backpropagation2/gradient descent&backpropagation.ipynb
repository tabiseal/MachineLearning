```python
import numpy as np
import matplotlib.pyplot as plt
import dataset
```

```python
xs,ys=dataset.get_beans(100)

#配置图像
plt.title("Size-Toxicity Function",fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")

plt.scatter(xs,ys)
w=0.1
y_pre=w*xs

plt.plot(xs,y_pre)

plt.show()
```

```python
for m in range(100):
    for i in range(100):
        x=xs[i]
        y=ys[i]
        #a=x^2
        #b=-2*x*y
        #c=y^2
        #斜率
        k=2*(x**2)*w+(-2*x*y)
        alpha=0.1
        w=w-alpha*k
#         plt.clf()#清空窗口
#         plt.scatter(xs,ys)
#         y_pre=2*xs
#         plt.xlim(0,1)
#         plt.ylim(0,1.2)
#         plt.plot(xs,y_pre)
#         plt.pause(0.01)
```
