```python
import numpy as np
import matplotlib.pyplot as plt
import dataset
```

```python
xs,ys=dataset.get_beans(100)
# print(xs)
# print(ys)
```

```python
plt.title("Size-Toxicity Function",fontsize=12)
plt.xlabel("bean Size")
plt.ylabel("Toxicity")

plt.scatter(xs,ys)

#y=0.5*x
w=0.5

y_pre=w*xs

# print(y_pre)
plt.plot(xs,y_pre)
plt.show()
```

```python
for m in range(100):
    for i in range(100):
        x=xs[i]
        y=ys[i]
        y_pre=w*x
        e=y-y_pre
        alpha=0.05
        w=w+alpha*e*x
    
y_pre=w*xs
print(y_pre)

plt.scatter(xs,ys)
plt.plot(xs,y_pre)
plt.show()
```
