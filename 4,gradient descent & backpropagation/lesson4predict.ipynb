```python
import numpy as np
import matplotlib.pyplot as plt
import dataset
```

```python
xs,ys=dataset.get_beans(100)

#配置图像
plt.title("Size-Toxicity Function",fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")
plt.xlim(0,1)
plt.ylim(0,1.2)

plt.scatter(xs,ys)

b=0.1
w=0.1

y_pre=w*xs+b

plt.plot(xs,y_pre)

plt.show()
```

```python
for m in range(500):
    for i in range(100):
        x=xs[i]
        y=ys[i]
        #a=x^2
        #b=-2*x*y
        #c=y^2
        #斜率
        dw=2*x**2*w+2*x*b-2*x*y
        db=2*b+2*x*w-2*y
        
        alpha=0.1
        
        w=w-alpha*dw
        b=b-alpha*db
        y_pre=w*xs+b
        
plt.scatter(xs,ys)
plt.plot(xs,y_pre)
plt.show()
```
